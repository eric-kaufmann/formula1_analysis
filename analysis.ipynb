## Merging Data

```python
constructor_results_df = pd.read_csv("data/constructorResults.csv")
constructor_results_df
```

```python
constructor_standings_df = pd.read_csv("data/constructorStandings.csv")
constructor_standings_df
```

```python
pd.merge(constructor_results_df, constructor_standings_df, on=['raceId', 'constructorId'], how='outer')
```

```python
results_df = pd.read_csv("data/results.csv")
results_df
```

```python
status_df = pd.read_csv("data/status.csv")
status_df
```

```python
pd.merge(results_df, status_df, on='statusId', how='outer')
```
